```python
def get_odds():
    for  number in range(10):
        if number % 2 != 0:
            yield number

#Initialize generator
odds = get_odds()

#Find a print the third value generated by the generator
count = 0
for value in odds:
    count += 1
    if count == 3:
        #print 3rd value
        print(value)
        break
```
